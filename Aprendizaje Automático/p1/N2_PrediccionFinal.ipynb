## Predicción con el modelo final
En este cuaderno vamos a utilizar el modelo final entrenadol, que hemos guardado previamente, para generar las predicciones sobre el conjunto de datos de evaluación (*bank_competition.pkl*).

Primero vamos a importar las librerías que vamos a necesitar.

```python
import pandas as pd
import joblib
```

A continuación, cargamos y preparamos los datos de evaluación. Para la preparación, tenemos en cuenta que hay que aplicar la misma lógica de preprocesamiento manual que usamos en el entrenamiento con la variable *pdays*, garantizando que el modelo reciba las columnas exactamente como las espera.

```python
# Cargamos los datos de test
df_test = pd.read_pickle('bank_competition.pkl') 

# Arreglamos la variable pdays
df_test['contactado_previamente'] = (df_test['pdays'] != -1).astype(int)
df_test.loc[df_test['pdays'] == -1, 'pdays'] = 0

print(f"Datos del examen cargados: {df_test.shape[0]} clientes listos para ser evaluados.")
```

También cargamos nuestro modelo ganador (Gradient Boosting) utilizando la librería *pickle* y le pasamos los datos nuevos para que genere las predicciones finales.

```python
# Cargamos nuestro modelo ganador
modelo = joblib.load('modelo_final.joblib')

# Le pedimos que realice las predicciones
predicciones = modelo.predict(df_test)

print("El modelo ha terminado de realizar las predicciones")
```

Finalmente, preparamos un DataFrame únicamente con el identificador del cliente y la predicción, y lo exportamos a formato *.csv*.

```python
# Creamos una tabla limpia
df_submission = pd.DataFrame({
    'Id': df_test.index,
    'Predicted': predicciones
})

# Lo guardamos en un csv
df_submission.to_csv('predicciones.csv', index=False)

print("Archivo 'predicciones.csv' generado correctamente")
```
